```python
# Write a program to count the number of words in a string without split()

s=input('enter the string')
L = []
temp= ' '
for i in s:

  if i != ' ':
    temp=temp+i
  else:
    L.append(temp)
    temp= ' '
L.append(temp)
print(L)
```

```python
# Write a python program to convert a string to title case without using a title()

s=input('enter the string')

L=[]
for i in s.split():
  L.append(i[0].upper()+i[1:].lower())
print(" ".join(L))
```

```python
# Write a program that can convert an integer to string

number=int(input('enter the number'))

digits= '0123456789'
result=  ''
while number !=0:
  result = digits [ number % 10]+ result
  number = number//10

print(result)
print(type(result))
```
