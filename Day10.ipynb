```python
#To reverse a input string

s=input('Enter a string')
reversed_string=s[::-1]
print(reversed_string)
```

```python
#count vowels
def count_vowels(string):
    vowels='aeiouAEIOU'
    count=0
    for char in string:
      if char in vowels:
        count +=1
    return count

s=input('Enter a string')
vowel_count=count_vowels(s)
print(vowel_count)
```
