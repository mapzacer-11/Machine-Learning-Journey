```python
#To reverse a input string

s=input('Enter a string')
reversed_string=s[::-1]
print(reversed_string)
```

```python
#count vowels
def count_vowels(string):
    vowels='aeiouAEIOU'
    count=0
    for char in string:
      if char in vowels:
        count +=1
    return count

s=input('Enter a string')
vowel_count=count_vowels(s)
print(vowel_count)
```

```python
#To remove spaces from string
def remove_spaces(string):
    return string.replace(' ','')
s=input('Enter a string')
new_string=remove_spaces(s)
print(new_string)
```

```python
#to check if string is palindrome or not
s=input('Enter a string')
if s==s[::-1]:
    print('Palindrome')
else:
    print('Not a palindrome')
```
