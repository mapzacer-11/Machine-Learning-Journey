```python
#Arithmetic Operators
print(7+9)
print(6-2)
print(3*3)
print(4/2)
print(5//2)
```

```python
#Relational Operators
print(4>5)
print(4<5)
print(4>=5)
print(4<=5)
print(4==4)
print(4!=4)
```

```python
#Logical Operators
print(1 and 0)
print(1 or 0)
print(not 1)
```

```python
#Bitwise Operators

#bitwise and
print(2&3)

#bitwise or
print(2|3)

#bitwise xor
print(2^3)

#bitwise nor
print(~3)

#left shift
print(4>>2)

#right shift
print(5<<2)
```

```python
#Assignment Operators
#a=a+2
a=2
a+=2
print(a)
```

```python
#Membership operators
#in/not in
print('A' in 'Adnan')
print('T'not in 'Tumkur')
```
