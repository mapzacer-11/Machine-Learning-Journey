```python
#problem

#sequence sum

n=int(input('enter n'))
result=0
fact=1
for i in range(1,n+1):
  fact=fact*1
  result=result+i/fact
  print(result)
```

```python
#Nested Loops
#unique pairs

for i in range(1,5):
  for j in range(1,5):
    print(i,j)
```

```python
#Questions

rows=int(input('enter number of rows'))
for i in range(1,rows+1):
  for j in range(1,i+1):
    print('*',end=' ')
  print()
```

```python
rows=int(input('enter number of rows'))
for i in range(1,rows+1):
  for j in range(1,i+1):
   print(j,end='')
  for k in range(i-1,0,-1):
    print(k,end='')
  print()
```

```python
#Loop Control Statements
#Break Statement
for i in range(1,5):
  if i==5:
    break
    print(i)
```

```python
lower=int(input('enter lower range'))
upper=int(input('enter upper range'))
for i in range(lower,upper+1):
  if i > 1:
    for j in range(2,i):
      if i%j==0:
        break
    else:
      print(i)
```

```python
#Continue Statement

for i in range(1,10):
  if i==5:
    continue
    print(i)
```
